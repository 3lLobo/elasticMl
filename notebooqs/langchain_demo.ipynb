```python
import pandas as pd
from pandasai import PandasAI
from langchain import HuggingFacePipeline
import torch

# Sample DataFrame
df = pd.DataFrame({
    "country": ["United States", "United Kingdom", "France", "Germany", "Italy", "Spain", "Canada", "Australia", "Japan", "China"],
    "gdp": [19294482071552, 2891615567872, 2411255037952, 3435817336832, 1745433788416, 1181205135360, 1607402389504, 1490967855104, 4380756541440, 14631844184064],
    "happiness_index": [6.94, 7.16, 6.66, 7.07, 6.38, 6.4, 7.23, 7.22, 5.87, 5.12]
})
```

```python
from transformers import LlamaTokenizer
```

```python
from transformers import AutoTokenizer
```

```python


n_params =  7
# model_path :  f"openlm-research/open_llama_{n_params}b"
model_path = "tiiuae/falcon-7b"
tokenizer = AutoTokenizer.from_pretrained(model_path)
#  GPT neo
# model_path :  "EleutherAI/gpt-neo-2.7B"

llm =  HuggingFacePipeline.from_model_id(
    # model_id="bigscience/bloom-1b7",
    model_id=model_path,
    task="text-generation",
    pipeline_kwargs={
        "max_new_tokens": 3000,
        "eos_token_id": tokenizer.eos_token_id,
        "do_sample": True,
        "top_k": 10,
        "num_return_sequences": 1,
    },
    model_kwargs={
        "temperature": 0, 
        # "max_length": 30000,
        # "max_new_tokens": 300,
        "early_stopping": True,
        # "max_time": 30,
        "max_length": 200,
        "do_sample": True,
        "top_k": 10,
        "num_return_sequences": 1,
        "eos_token_id": tokenizer.eos_token_id,
        },
    device=0,
)
```

```python
res = llm.predict("The United States is a country that is", num_return_sequences=1, max_length=1000, max_new_tokens=300, max_time=30)
print(res)
```

```python
from langchain import PromptTemplate, LLMChain

template = """This is my complete, step-by-step recipe for a delicious {dish}: 
First """
prompt = PromptTemplate(template=template, input_variables=["dish"])

llm_chain = LLMChain(prompt=prompt, llm=llm, verbose=True)

# question = "What is electroencephalography?"
dish = "death-by-chocolate cake"

print(llm_chain.run(dish))
```

```python
from langchain.agents import create_pandas_dataframe_agent
from langchain.agents.agent_types import AgentType

agent = create_pandas_dataframe_agent(
    llm,
    df,
    verbose=True,
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION
)
```

```python
agent.run("how many rows are there?")
```

```python
agent.run("Which are the 5 happiest countries?")
```

```python


# Instantiate a LLM
from pandasai.llm.langchain import LangchainLLM

llm = LangchainLLM(llm)

# llm = OpenAI(api_token="YOUR_API_TOKEN")

pandas_ai = PandasAI(llm)
res = pandas_ai(df, prompt='Which are the 5 happiest countries?')
print(res)
```
